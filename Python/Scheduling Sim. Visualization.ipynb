```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

<h1>Scheduling Simulator Basic Visualizer</h1>
This script is meant to pair with my CPU scheduling test generator & simulator in order to provide a simple analysis of the results it provides.

<h3>Import CSVs</h3>
Ensure these are in the same folder as the python notebook.

```python
fcfs = pd.read_csv('Results_fcfs.csv', low_memory=False)
rr = pd.read_csv('Results_rr.csv', low_memory=False)
sjf = pd.read_csv('Results_sjf.csv', low_memory=False)
srtn = pd.read_csv('Results_srtn.csv', low_memory=False)
```

<h4>Group and Average</h4>
Next, we group by program count and get average values to simplify results into something readable

```python
fcfsAvgs = fcfs.groupby('Program Count').mean()
rrAvgs = rr.groupby('Program Count').mean()
sjfAvgs = sjf.groupby('Program Count').mean()
srtnAvgs = srtn.groupby('Program Count').mean()
```

<h4>Plotting Basic Results</h4>

```python
fig = plt.figure(figsize = (15,38))
grid = plt.GridSpec(5,1, hspace = 0.5, wspace = 0.3, figure = fig)
linbin = np.arange(5, 11, 1)
ticks = range(5, 11, 1)

fig.add_subplot(grid[1,0], title = "Average Turn-Around Time", xlabel = 'Program Count', ylabel = 'Time Units')
plt.xticks(ticks)
plt.plot(linbin, fcfsAvgs['Avg. Turnaround Time'], label = 'FCFS')
plt.plot(linbin, rrAvgs['Avg. Turnaround Time'], label = 'RR')
plt.plot(linbin, sjfAvgs['Avg. Turnaround Time'], label = 'SJF')
plt.plot(linbin, srtnAvgs['Avg. Turnaround Time'], label = 'SRTN')
plt.legend()

fig.add_subplot(grid[2,0], title = "Average Wait Time", xlabel = 'Program Count', ylabel = 'Time Units')
plt.xticks(ticks)
plt.plot(linbin, fcfsAvgs['Avg. Wait Time'], label = 'FCFS')
plt.plot(linbin, rrAvgs['Avg. Wait Time'], label = 'RR')
plt.plot(linbin, sjfAvgs['Avg. Wait Time'], label = 'SJF')
plt.plot(linbin, srtnAvgs['Avg. Wait Time'], label = 'SRTN')
plt.legend()

fig.add_subplot(grid[3,0], title = "Average Response Time", xlabel = 'Program Count', ylabel = 'Time Units')
plt.xticks(ticks)
plt.plot(linbin, fcfsAvgs['Avg. Response Time'], label = 'FCFS')
plt.plot(linbin, rrAvgs['Avg. Response Time'], label = 'RR')
plt.plot(linbin, sjfAvgs['Avg. Response Time'], label = 'SJF')
plt.plot(linbin, srtnAvgs['Avg. Response Time'], label = 'SRTN')
plt.legend()

fig.add_subplot(grid[4,0], title = "Average Completion Time", xlabel = 'Program Count', ylabel = 'Time Units')
plt.xticks(ticks)
plt.plot(linbin, fcfsAvgs['Avg. Completion Time'], label = 'FCFS')
plt.plot(linbin, rrAvgs['Avg. Completion Time'], label = 'RR')
plt.plot(linbin, sjfAvgs['Avg. Completion Time'], label = 'SJF')
plt.plot(linbin, srtnAvgs['Avg. Completion Time'], label = 'SRTN')
plt.legend()
```
